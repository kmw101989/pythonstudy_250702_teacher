```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
from openpyxl import Workbook

# 1.ì…€ë ˆë‹ˆì›€ ê¸°ë³¸ ì˜µì…˜ ì„¤ì •
options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--headless")

driver = webdriver.Chrome(options = options)

# 2.í˜ì´ì§€ ì ‘ì†
url = "https://www.kurly.com/goods/1000078028"
driver.get(url)

time.sleep(5) # í˜ì´ì§€ ë¡œë”©ëŒ€ê¸° ì‹œê°„ ì„¤ì •

# 3. HTML ê°€ì ¸ì˜¤ê¸°
soup = BeautifulSoup(driver.page_source, "html.parser")

# 4. ë¦¬ë·° í…ìŠ¤íŠ¸ ì¶”ì¶œí•˜ê¸°
reviews = soup.select("p.css-y49dcn.e36z05c13")
review_lists = [review.get_text().strip() for review in reviews]

driver.quit()

wb = Workbook()
ws = wb.active
ws.title = "Kurly Reviews"

ws.append(["ë¦¬ë·° ë‚´ìš©"])
for item in review_lists :
    ws.append([item])

wb.save("kurly_reviews.xlsx")
print("ì—‘ì…€ ì €ì¥ ì™„ë£Œ : kurly_reviews.xlsx")
```
