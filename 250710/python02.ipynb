```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/")
soup = BeautifulSoup(res.content, "html.parser")

trs = soup.select("tr")

# for tr in trs :
#     tds = tr.select("td")
#     for td in tds :
#         print(td.get_text())

for tr in trs :
    tds = tr.find_all("td")
    row_str = ""
    for td in tds :
        row_str += f" / {td.get_text()}"
    print(row_str[2:])
```
